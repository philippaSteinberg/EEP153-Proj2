### Demographic Stats Oakland Charter Academy (2021)

```python
import pandas as pd
import numpy as np
import pylab
```

```python
# STATS from OCA
budget_stats = {
    "Section": ["School Nutrition", "Mental Health"],
    "Total Budgeted Funds $": [10000, 25000],
    "Estimated Actual Expenditures $": [13708, 28530]
    }
enrollment_stats = {
    "Grade": ["year 7", "year 8", "year 9"],
    "Number of Students": [81,81,84]
    }

demography_stats = {
    "Group": ["Female", "Male", "Non-Binary", "American Indian or Alaska Native", "Asian", "Black or African American", "Hispanic or Latino", "Two or More Races", "White", "English Learners", "Socioeconomically Disadvanted", "Not Socioeconomically Disadvanted", "Students with Disabilities"],
    "Percent Students": [55, 45, 0, 0.82, 1.23, 2.47, 88.9, 0.41, 0.41, 5.76,53, 47, 9]
    }

lunch_stats = {
    "Status": ["Eligible For Free Lunch", "Eligible For Reduced Lunch"],
    "Percent Students": [73,7]
    }
```

```python
OCA_Spendings = pd.DataFrame(budget_stats)
OCA_Enrollment = pd.DataFrame(enrollment_stats)
OCA_Demography = pd.DataFrame(demography_stats).set_index("Group")
OCA_Lunch = pd.DataFrame(lunch_stats).set_index("Status")
```

```python
plot_spendings = OCA_Spendings.plot.bar(
        x= "Section",
        y= ["Total Budgeted Funds $", "Estimated Actual Expenditures $"],
        rot = 1,
        title = "Oakland Charter Academy Budget vs Spending", 
        figsize=(11, 6), 
        color=("lightblue", "mediumslateblue")
        )
```

```python
plot_enrollment = OCA_Enrollment.plot.bar(
        x= "Grade",
        y= "Number of Students",
        rot = 1,
        title = "Oakland Charter Academy Enrollment by Grade",
        figsize=(11, 6),
        color=("lightblue")
        )
```
